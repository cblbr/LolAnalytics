```
def predict_probability_of_winning(gold_diff_at_10, exp_diff_at_10):
  
  from joblib import load
  import numpy as np
  import pandas as pd
  
  
  model = load('model.joblib.dat')
  dictionary = {'blueGoldDiff':[gold_diff_at_10],'blueExperienceDiff':[exp_diff_at_10]}
  
  df = pd.DataFrame(dictionary)

  predicts = model.predict_proba(df)

  cols2 = ['probability lose','probability win']
  for i,col in enumerate(cols2):
    df[col] = predicts[:, i]
  return df





```

```
predict_probability_of_winning(5000,3000)
```
